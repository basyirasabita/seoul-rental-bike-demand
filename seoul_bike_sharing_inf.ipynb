# Import Libraries

```python
# Import Libraries
import pandas as pd
import pickle
import numpy as np
```

# Load Model

```python
# Load Model
with open('deployment/model_gbm_reg.pkl', 'rb') as file_1:
  model = pickle.load(file_1)

model
```

# Making Dummy Data

```python
# Making dummy data
data_inf = {
    'Date': ['2023-01-15', '2023-04-17', '2023-06-19', '2023-10-31'], 
    'Hour': [9, 13, 17, 20],
    'Temperature': [2.08, 20.3, 28, 9], 
    'Humidity': [78, 26, 51, 23],
    'Wind Speed': [4, 2.5, 2.4, 3], 
    'Visibility': [275, 1134, 1705, 1000], 
    'Dew Point Temperature': [-11.84, -0.8, 15, 2], 
    'Solar Radiation': [0.11, 3.16, 3.8, 0.7],
    'Rainfall': [0, 0, 0, 0], 
    'Snowfall': [0.12, 0, 0, 0], 
    'Seasons': ['Winter', 'Spring', 'Summer', 'Autumn'], 
    'Holiday': ['No Holiday', 'No Holiday', 'No Holiday', 'No Holiday'], 
    'Functioning Day': ['Yes', 'Yes', 'Yes', 'Yes']
}

df_inf = pd.DataFrame(data_inf)
df_inf
```

# Predict the Rented Bike Count

```python
# Predict the inference data
y_pred_inf = model.predict(df_inf)

print(f'Predicted Bike Rented Count on {df_inf.Date.loc[0]} ({df_inf.Seasons.loc[0]}): {round(y_pred_inf[0], 0)}')
print(f'Predicted Bike Rented Count on {df_inf.Date.loc[1]} ({df_inf.Seasons.loc[1]}): {round(y_pred_inf[1], 0)}')
print(f'Predicted Bike Rented Count on {df_inf.Date.loc[2]} ({df_inf.Seasons.loc[2]}): {round(y_pred_inf[2], 0)}')
print(f'Predicted Bike Rented Count on {df_inf.Date.loc[3]} ({df_inf.Seasons.loc[3]}): {round(y_pred_inf[3], 0)}')
```

We've successfully predict the Bike Rented Count for each seasons. The predicted values would have an error of ±172 bike rental demands.
